Q. 1. Explain the Different types of data ( qualitative and quantitative ) and provide examples of each. Discuss, nominal, ordinal interval, and ratio scales.

Answer:-

Qualitative data (categorical data): Describes characteristics or qualities. Example: colors (red, blue), types of cuisine.
Quantitative data (numerical data): Represents amounts or quantities. Example: height (cm), temperature (°C).
Nominal scale: Categories with no specific order. Example: gender, nationality.
Ordinal scale: Categories with a meaningful order. Example: satisfaction levels (low, medium, high).
Interval scale: Numeric values with no true zero. Example: temperature in Celsius.
Ratio scale: Numeric values with a true zero. Example: weight, height.

Q. 2. What are the Measures of central tendency, and when should you use each? Discuss the mean, median, and mode with examples and situations where each is appropriate.

Answer :-

Mean: The average value. Used when data is symmetrical without outliers.
Median: The middle value. Used for skewed data or when outliers are present.
Mode: The most frequent value. Used when data has distinct peaks (e.g., categorical data).
Example:
A dataset: 2, 3, 5, 5, 7
Mean: (2 + 3 + 5 + 5 + 7) / 5 = 4.4
Median: 5
Mode: 5

Q. 3.  Examples the concept of dispersion. Ho do variance and standard deviation measure the spread of data?

Answer :-

Measures how spread out the data is.
Variance: The average of squared differences from the mean. Measures data spread.
Standard deviation: The square root of variance, indicating how much data deviates from the mean.

Q. 4.  What is Box plot, and what can it tell you about the distribution of data?

Answer:-

A graphical representation showing the minimum, first quartile (Q1), median, third quartile (Q3), and maximum. It helps identify:
The spread of data.
Outliers.
Skewness (based on the position of the median and quartiles).

Q. 5. Discuss the role of Random sampling in making inferences about populations.

Answer:-

A technique used to select a representative subset of a population for analysis.
It ensures unbiased, generalizable conclusions about the entire population.

Q.6.Explain the concept of Skewness and its types. How does skewness affect the interpretation of data?

Answer:-

Describes asymmetry in a dataset.
Positive skew: Tail on the right (mean > median).
Negative skew: Tail on the left (mean < median).
Skewness affects interpretation by showing where most values are concentrated.

Q.7.What is the Interquartile range (IQR) And how is it used do detect outliers?

Answer:-

The range between the first quartile (Q1) and the third quartile (Q3). Used to detect outliers as values below Q1 - 1.5 * IQR or above Q3 + 1.5 * IQR.

Q.8. Discuss the conditions under which the binomial distribution is used.

Answer:-

Used when:
There are a fixed number of trials.
Each trial has two possible outcomes (success/failure).
Probability of success remains constant.
Trials are independent.

Q.9 Explain the properties of the normal distribution and the empirical rule (68-95-99.7 rule).

Answer:-

A bell-shaped curve symmetric about the mean.
The empirical rule (68-95-99.7 rule):
68% of data lies within 1 standard deviation of the mean.
95% lies within 2 standard deviations.
99.7% lies within 3 standard deviations.

Q.11. Explain what a Random variable is and differentiate between discrete and continuous random variables.

Answer:-

A variable representing outcomes of a random phenomenon.
Discrete random variable: Takes specific values (e.g., number of heads in a coin toss).
Continuous random variable: Takes any value within a range (e.g., temperature).

Q.12. Provide an example dataset, Calculate both Covariance and correlation, and interpret the results.

Answer:-

Covariance: Measures how two variables vary together.
Correlation: A standardized measure of the relationship between two variables (ranges from -1 to 1). Example dataset: (X: 2, 3, 4; Y: 5, 6, 8)
Covariance: Calculate mean of X and Y, find deviations, and compute the average product of deviations.
Correlation: Normalize covariance by dividing by the product of standard deviations of X and Y.
